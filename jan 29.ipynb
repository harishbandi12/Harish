```python
import datetime
f=open("data.txt")
s=f.read()
l=eval(s)
print(type(l))
s=set()
s1=set()
for i in l:
	s.add(i['student_id'])
	s1.add(i['date'])	
s1=list(s1)
s1.sort()

d={}
for i in l:
	if(i['date']==s1[0]):
		if(i['student_id'] not in list(d.keys())):
			d[i['student_id']]=list()
			d[i['student_id']].append(i['attendance'])
		else:
			d[i['student_id']].append(i['attendance'])
	if(i['date']==s1[1]):
		if(i['student_id'] not in list(d.keys())):
			d[i['student_id']]=list()
			d[i['student_id']].append(i['attendance'])
		else:
			d[i['student_id']].append(i['attendance'])
	if(i['date']==s1[2]):
		if(i['student_id'] not in list(d.keys())):
			d[i['student_id']]=list()
			d[i['student_id']].append(i['attendance'])
		else:
			d[i['student_id']].append(i['attendance'])
for i in d.keys():
	if(len(d[i])<6):
		d[i].insert(2,"NA")
	print(d[i])
	
```
